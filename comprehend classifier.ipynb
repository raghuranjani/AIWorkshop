```python
import os
```

```python
import pandas as pd
```

```python
df = pd.DataFrame()
mapping = {}
source_path = "bbc/bbc/"
```

```python
os.listdir(source_path)
```

```python
for i in sorted(os.listdir(source_path)):
    if os.path.isdir(source_path+i):
        print (i)
        mapping[i] = sorted(os.listdir(source_path+i))[:300]
        print (mapping[i])
```

```python
mapping.keys()
```

```python
mapping.items()
```

```python
label = []
data = []

for i,j in mapping.items():
    for k in j:
        label.append(i)
        data.append(open(source_path+i+"/"+k,encoding="cp1252").read().replace("\n", " "))
```

```python
label
```

```python
data[:10]
```

```python
df["label"] = label
```

```python
df["data"] = data
```

```python
df.head()
```

```python
df = df.sample(frac=1).reset_index(drop=True)
```

```python
df.to_csv("train.csv",index=False,header=False)
```
